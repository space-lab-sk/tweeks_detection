## Vykreslenie anotácií

### Vstupy
data: cesta k súboru anotácií makers-classifications2.csv<br>
rootPath: cesta k spektrogramom bez anotácií<br>
csvOut: cesta, kde bude uložený výstupný súbor

```python
data=pd.read_csv("files/markers-classifications2.csv")
rootPath = r"yolo/data/image"
csvOut='files/annot_output.csv'

data=data.drop(data.index[0:50])
a=50

with open(csvOut, mode='w') as csv_file:
    fieldnames = ['image','xmin', 'ymin', 'xmax', 'ymax', 'label']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    while (a<=len(data)-1):
        load=json.loads(data.iloc[a, 11])[0]["value"]
        filesy=json.loads(data.iloc[a,12])[str(data.iloc[a,13])]["Filename"].split('.')
        filesy=filesy[0]+'.jpg'
        for k in load:
            writer.writerow({'image': filesy,'xmin': round(k["x"]),'ymin': 0,'xmax': round(k["x"]+k["width"]),  
            'ymax': 162,'label': k["tool_label"]})
            
        for root, dirs, files in os.walk(rootPath):
            for file in files:
                if file == filesy:
                    im = Image.open(os.path.join(root, "%s" % (file)))
                    w, h = im.size
                    im1 = ImageDraw.Draw(im)
                    num=0
                    for k in load:
                        if (k["tool_label"] == 'Blesk'):
                            shape=[(k["x"], 0), (k["x"], h)]
                            im1.line(shape, fill="black", width=2)
                            shape=[(k["x"]+k["width"], 0), (k["x"]+k["width"], 162)]
                            im1.line(shape, fill="black", width=2)
                        else:
                            shape=[(k["x"], 0), (k["x"], h)]
                            im1.line(shape, fill="white", width=2)
                            shape=[(k["x"]+k["width"], 0), (k["x"]+k["width"], 162)]
                            im1.line(shape, fill="white", width=2)
                        num+=1
                    print("File %s done. Written %d events." % (filesy, num))
                    im.save(os.path.join('files/annotations', "1%s" % (file)))
        a+=1

```

## Vytvorenie výstupnej tabuľky

### Vstupy
data: cesta k súboru anotácií makers-classifications2.csv<br>
csvOut: cesta, kde bude uložený výstupný súbor

```python
data=pd.read_csv("files/markers-classifications2.csv")
csvOut='files/annot_output2.csv'

data=data.drop(data.index[0:50])
numOfTweek=0
with open(csvOut, mode='w') as csv_file:
    fieldnames = ['event','signal', 'date', 'second', 'milisecond', 'tweek signs', 'f_signs [kHz]', 'f_min<2kHz']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()

    a=0
    while (a<=len(data)-1):
        filename=json.loads(data.iloc[a,12])[str(data.iloc[a,13])]["Filename"]
        value=json.loads(data.iloc[a, 11])[0]["value"]
        if (json.loads(data.iloc[a, 10])["subject_dimensions"][0]):
            widthms=(json.loads(data.iloc[a, 10])["subject_dimensions"][0]["naturalWidth"])/1000
        event=a+1
        date=filename[4:19]
        sec=filename[filename.find("s")+1:filename.find(".png")]
        sig=0
        for k in value:
            milsec=round((((k["width"])/2)+k["x"])/widthms)
            if len(k["details"])>0:
                twsigns=k["details"][0]["value"]
            else: twsigns=None
            sig+=1
            numOfTweek+=1

            writer.writerow({'event': event,'signal':sig, 'date': date, 'second': sec, 'milisecond': milsec, 'tweek signs': twsigns, 
                'f_signs [kHz]': None, 'f_min<2kHz': None})
        a+=1
```

## Normalizovanie anotácií na vstup pre neurónovú sieť YOLO

### Vstupy
data: cesta k súboru anotácií makers-classifications2.csv<br>
rootPath: cesta k spektrogramom bez anotácií<br>
trainPics: cesta k množine obrázkov, pre ktoré budú vytvorené normalizované súbory

```python
data=pd.read_csv("files/markers-classifications2.csv")
rootPath = r"/image"
trainPics="./files/image"
textFile=trainPics + "/txt"

data=data.drop(data.index[0:50])
a=50

while (a<=len(data)-1):
    load=json.loads(data.iloc[a, 11])[0]["value"]
    filesy=json.loads(data.iloc[a,12])[str(data.iloc[a,13])]["Filename"].split(".")

    with open(os.path.join(textFile, "%s.txt" % (filesy[0])), mode='w') as csv_file:
        fieldnames1 = ['label','x', 'y', 'width', 'height']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames1, delimiter=' ')
        # writer.writeheader()
        for k in load:
            writer.writerow({'label':0, 'x':round((((k["x"])+(k["width"]/2))/1421),5), 'y':0.5, 'width':round((k["width"]/1421),5), 'height':1})                                             #height

    print("Just made normalized txt file for %s" % (filesy[0]))
    a+=1
```
